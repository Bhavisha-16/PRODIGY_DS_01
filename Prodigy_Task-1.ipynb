```python
import matplotlib.pyplot as plt

# Sample data
age_groups = ['0-14 years', '15-64 years', '65+ years']
population_percentages = [25.8, 65.4, 8.8]  # These are hypothetical values for illustration

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(age_groups, population_percentages, color=['blue', 'green', 'red'])

# Add titles and labels
plt.title('World Population Distribution by Age Group (2022)')
plt.xlabel('Age Groups')
plt.ylabel('Percentage of Total Population')

# Display the chart
plt.show()
```
